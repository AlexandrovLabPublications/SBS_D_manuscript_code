```python
import os

files = os.listdir('/Users/kazachkova/Documents/tscc/restricted/topography_figures_MSS/nucleosome_occupancy_data/signaturebased/')
```

```python
files = pd.Series(files)

files2 = files[files.str.contains('SBS288B')]
files3 = files2[files2.str.contains('AverageSignalArray.txt')]
files4 = files3[files3.str.contains('sim')]
files4
```

```python
df_288b = pd.DataFrame(index=list(range(1,101)),
                       columns=list(range(0,2001)),dtype=float)
i=1
for file in files4:
   
    sim = pd.read_table('/Users/kazachkova/Documents/tscc/restricted/topography_figures_MSS/nucleosome_occupancy_data/signaturebased/' + file,
                 header=None).T
    df_288b.loc[i] = sim.loc[:,0]
    i+=1
```

```python
df_288b_mean_sim = df_288b.mean()

df_288b_mean_sim.index = df_288b_mean_sim.index - 1000

average_signal_array_SBS288B.index = average_signal_array_SBS288B.index - 1000
```

```python
fig,ax = plt.subplots(figsize=(6,4))
plt.plot(df_288b_mean_sim)
plt.plot(average_signal_array_SBS288B.loc[:,0])
plt.title('SBS288B')

plt.ylim(.75,1.05)
```

```python
# same thing 288D
average_signal_array_SBS288D = pd.read_table('/Users/kazachkova/Documents/tscc/restricted/topography_figures_MSS/nucleosome_occupancy_data/signaturebased/SBS288D_AverageSignalArray.txt',
             header=None).T



files2 = files[files.str.contains('SBS288D')]
files3 = files2[files2.str.contains('AverageSignalArray.txt')]
files4 = files3[files3.str.contains('sim')]


files4

df_288d = pd.DataFrame(index=list(range(1,101)),
                       columns=list(range(0,2001)),dtype=float)
i=1
for file in files4:
   
    sim = pd.read_table('/Users/kazachkova/Documents/tscc/restricted/topography_figures_MSS/nucleosome_occupancy_data/signaturebased/' + file,
                 header=None).T
    df_288d.loc[i] = sim.loc[:,0]
    i+=1

    

df_288d_mean_sim = df_288d.mean()

df_288d_mean_sim.index = df_288d_mean_sim.index - 1000

average_signal_array_SBS288D.index = average_signal_array_SBS288D.index - 1000

fig,ax = plt.subplots(figsize=(6,4))
plt.plot(df_288d_mean_sim)
plt.plot(average_signal_array_SBS288D.loc[:,0])
plt.title('SBS288D')
plt.ylim(.75,1.05)
```

```python
# and we need SBS18 

sbs18_all_types = pd.read_table('/Users/kazachkova/Downloads/v3.2_SBS18_NUCLEOSOME_OCC_BURCAK.txt',skiprows=[0])
sbs18_all_types = sbs18_all_types.loc[0:12]
sbs18_all_types_df = pd.DataFrame()
i=0
for x in sbs18_all_types.loc[:,'real_average_signal_array']:
    x = x.replace('[','').replace(']','')
    x = x.split(',')
    sbs18_all_types_df[i] = x
    i+=1

    
sbs18_all_types_df_mean = sbs18_all_types_df.astype(float).mean(axis=1)



sbs18_all_types_df_sim = pd.DataFrame()
i=0
for x in sbs18_all_types.loc[:,'sims_average_signal_array']:
    x = x.replace('[','').replace(']','')
    x = x.split(',')
    sbs18_all_types_df_sim[i] = x
    i+=1


sbs18_all_types_df_sim_mean = sbs18_all_types_df_sim.astype(float).mean(axis=1)


sbs18_all_types_df_mean.index = sbs18_all_types_df_mean.index - 1000
sbs18_all_types_df_sim_mean.index = sbs18_all_types_df_sim_mean.index - 1000

plt.plot(sbs18_all_types_df_sim_mean)
plt.plot(sbs18_all_types_df_mean)
plt.ylim(.83,1.13)

plt.title('SBS18_all')
```

```python
# SBS18 just CRC 


sbs18_crc = sbs18_all_types[sbs18_all_types.loc[:,'cancer_type'] == 'ColoRect-AdenoCA'].loc[:,'real_average_signal_array']
sbs18_crc = sbs18_crc.str.replace('[','').str.replace(']','')
sbs18_crc = pd.Series(sbs18_crc.str.split(',')[1]).astype(float)


sbs18_crc_sim = sbs18_all_types[sbs18_all_types.loc[:,'cancer_type'] == 'ColoRect-AdenoCA'].loc[:,'sims_average_signal_array']
sbs18_crc_sim = sbs18_crc_sim.str.replace('[','').str.replace(']','')
sbs18_crc_sim = pd.Series(sbs18_crc_sim.str.split(',')[1]).astype(float)


sbs18_crc.index = sbs18_crc.index - 1000
sbs18_crc_sim.index = sbs18_crc_sim.index - 1000

plt.plot(sbs18_crc)
plt.plot(sbs18_crc_sim)
plt.ylim(.83,1.13)

plt.title('SBS18_CRC')
```

```python
############### Figure 2c ###############

fig,ax = plt.subplots(1,2,figsize=(14,4),sharey=True,constrained_layout=True)

ax[0].plot(df_288d_mean_sim,color='grey',linestyle='--')
ax[0].plot(average_signal_array_SBS288D.loc[:,0],color='mediumblue')
ax[0].axvline(0,color='lightgrey',linestyle=':')
ax[0].set_title('SBS_D')
# ax[0].set_xticks([])

ax[1].plot(df_288b_mean_sim,color='grey',linestyle='--',label='Simulated SBS')
ax[1].plot(average_signal_array_SBS288B.loc[:,0],color='mediumblue',label='Real SBS')
ax[1].axvline(0,color='lightgrey',linestyle=':')
ax[1].set_title('SBS18^')
ax[1].legend(frameon=False,fontsize=14)
# ax[1].set_xticks([])


ax[0].locator_params(axis='x', nbins=5)

ax[1].locator_params(axis='x', nbins=5)

ax[0].set_ylim(.75,1.04)

for i in range(0,2):
#     for j in range(0,2):
    ax[i].spines['top'].set_visible(False)
    ax[i].spines['right'].set_visible(False)

    for item in ([ax[i].title, ax[i].xaxis.label, ax[i].yaxis.label] +
             ax[i].get_xticklabels() + ax[i].get_yticklabels()):
        item.set_fontsize(14)

        
fig.supxlabel('Interval around mutation (bp)', fontsize=14)
fig.supylabel('Average Nucleosome Signal', fontsize=14)


# plt.tight_layout()

# plt.savefig('plots_for_paper/topography_TOP.png',dpi=300)
```
