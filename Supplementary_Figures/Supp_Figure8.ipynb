```python
# TCGA - ALL

results = pd.read_csv('/Users/kazachkova/Documents/alexandrov_lab/mutographs/CRC_clonal_analysis/V3_Data/examine_SBS288D/Reviewers_Comments/Reviewers_comments_V2/TCGA_gene_expression_work/res_limma.csv',index_col=0)

for_mapping = pd.read_table('/Users/kazachkova/Documents/alexandrov_lab/mutographs/CRC_clonal_analysis/V3_Data/data_from_other_papers/TCGA_data_files/gene_expression/gdc_download_20240814_205741.933713/00ae9ab8-6eaa-4085-af72-26f96df97fa3/90c9f8cd-4c8c-4f07-af2f-e17db69bd561.rna_seq.augmented_star_gene_counts.tsv',skiprows=[0,1,2,3,4,5],header=None,index_col=0).loc[:,1]


results

results.index = for_mapping.loc[results.index]

results = results.dropna()

import numpy as np
results['-log10(qval)'] = -np.log10(results.loc[:,'adj.P.Val'])

results

results['status'] = 'ns'

results.loc[results[results.loc[:,'logFC']> .25].index & 
           results[results.loc[:,'adj.P.Val']<0.05].index,'status'] = 'up-regulated'

results.loc[results[results.loc[:,'logFC']< -.25].index & 
           results[results.loc[:,'adj.P.Val']<0.05].index,'status'] = 'down-regulated'

results = results.sort_values('status')

fig,ax = plt.subplots(figsize=(6,6))
sns.scatterplot(data=results,
               x='logFC',y='-log10(qval)',hue='status',palette = ['blue','grey','red'])

plt.xlim(-1.1,1.1)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.legend(frameon=False,title='',fontsize=14,
          loc='upper center', bbox_to_anchor=(0.5, -0.15),
          ncol=3)

for item in ([ax.title, ax.xaxis.label, ax.yaxis.label] +
             ax.get_xticklabels() + ax.get_yticklabels()):
    item.set_fontsize(14)
    
plt.axvline(-.25,color='black',linestyle='--')
plt.axvline(.25,color='black',linestyle='--')

plt.axhline(-np.log10(0.05),color='black',linestyle='--')
plt.title('TCGA (all SBS_D+ samples)',fontsize=14)
# plt.savefig('plots_for_responses/TCGA_expression_all.png',dpi=300)
```
